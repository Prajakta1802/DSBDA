Perform the following operation using Python on the Facebook metrics data sets:


1.   Create data subsets
2.   Merge Data
3.   Sort Data
4.   Transporting Data
5.   Shape and Reshape Data



```python
#Importing Required Libraries

import pandas as pd
import numpy as np
```

```python
#Reading dataset

facebook = pd.read_csv('/content/dataset_Facebook.csv', ";")
facebook
```

## **Creating Subset**

```python
#Creating photo subset

photos = facebook[facebook["Type"] == "Photo"]
print("No. of Records available for Photos : ", len(photos))
```

```python
#creating video subset

videos = facebook[facebook["Type"] == "Video"]
videos
print("No. of Records available for Vidoes : ", len(videos))
```

```python
#creating links subset

links = facebook[facebook["Type"] == "Link"]
links
print("No. of Records available for Links : ", len(links))
```

```python
#Creating status subset

status = facebook[facebook["Type"] == "Status"]
status
print("No. of Records available for Status : ", len(status))
```

## **Merge Dataset**

```python
#Merging the dataset using concat() function

merged_dataset = pd.concat([links, videos])
merged_dataset
```

```python
print("After merging no of records available: ", len(merged_dataset))
```

## **Sort Data**

```python
#Sorting dataset using values of Total interactions

photos.sort_values('Total Interactions')
```

## **Transpose Data**

```python
#Transposing dataset 

result = videos.transpose()
result
```

## **Shape and Reshape Data**

```python
#Displaying Shape of every dataset

print("Shape of Videos dataset: ",videos.shape)
print("Shape of Photos datset: ",photos.shape)
print("Shape of Links dataset:",links.shape)
print("Shape of Status dataset:",status.shape)
```

```python
videos
```

```python
#Melting videos subset by selecting two columns

melted = pd.melt(videos, id_vars = ["Page total likes"], value_vars=["Post Month", "Post Weekday"], var_name = "Attribute", value_name = "Value")
melted
```

```python
#Transforming back to it's original form 

melted.pivot_table(index="Page total likes", columns="Attribute", values="Value", aggfunc='first')
```

```python
#Creating another subset for melting having unique elements

melt_input_subset = videos[['Page total likes', 'Lifetime Post Total Reach', 'Lifetime Post Total Impressions', 'Lifetime Engaged Users', 'Lifetime Post Consumers', 'Lifetime Post Consumptions']]
melt_input_subset
```

```python
print("Before melting shape of dataset: ",melt_input_subset.shape)
print("Before melting no of records available: ", len(melt_input_subset))
```

```python
#Melting the subset

melted_videos = pd.melt(melt_input_subset, id_vars = ["Page total likes"], value_vars=['Lifetime Post Total Reach', 'Lifetime Post Total Impressions', 'Lifetime Engaged Users', 'Lifetime Post Consumers', 'Lifetime Post Consumptions'], var_name = "Attribute", value_name = "Value")
melted_videos
```

```python
print("After melting shape of dataset: ",melted_videos.shape)
print("After melting no of records available: ", len(melted_videos))
```

```python
#Transforming it to back uisng pivot function

original_form = melted_videos.pivot_table(index="Page total likes", columns="Attribute", values="Value", aggfunc='first')
original_form
```
